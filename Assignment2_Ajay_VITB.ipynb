```python
import seaborn as sns
```

```python
df1 = sns.load_dataset('car_crashes')
```

```python
df1
```

```python
df1.info()
```

```python
sns.scatterplot(x="total",y="alcohol",data=df1)
```

```python
corr = df1.corr()
corr
```

```python
sns.heatmap(corr, annot=True)
```

**Inference** = from above we get that we can use which correlations for visualisation to get something usefull from that

Hence  we are going to use following relations for visulaisation

1.total vs speeding

2.alcohol vs speeding

3.alcohol vs not_distracted

4.not_distracted vs no_previous
